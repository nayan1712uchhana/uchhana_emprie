```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import datasets
import seaborn as sns 
```

```python
data = datasets.load_boston()
data.keys()
```

```python
df = pd.DataFrame(data=data['data'],columns=data['feature_names'])
```

```python
df.head(21)
```

```python
print(data['DESCR'])
```

```python
 data['MVD'] = data.target
```

```python
df.isnull().sum()
```

```python
sns.set(rc={'figure.figsize':(9,9.5)})
sns.distplot(data['MVD'])
plt.show()
```

```python
df.columns
```

```python
x = df[['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX',
       'PTRATIO', 'B', 'LSTAT']]
y = data['MVD']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.25)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
x_train.head(5)
```

```python
from sklearn.linear_model import LinearRegression
lg = LinearRegression()
lg.fit(x_train,y_train)
```

```python
pred  = lg.predict(x_test)
print(f'The predicted values by LinearRegression are: \n{pred}')
```

```python
sns.scatterplot(x=data['MVD'] , y=df['LSTAT'],hue = df['TAX'])
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test,pred)
mse
```

```python
import math
math.sqrt(mse)
```

```python
plt.figure(figsize=(20, 5))
for i, col in enumerate(df.columns):
    plt.subplot(1, 3, 2)
    x = df[col]
    y = data['MVD']
    plt.plot(x, y, '*')
    plt.plot(np.unique(x), np.poly1d(np.polyfit(x, y, 1))(np.unique(x)))
    plt.title(col)
    plt.xlabel(col)
    plt.ylabel('prices')
plt.tight_layout()
```
